```python
def train_one_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss = 0.0
    n_batches = 0
    for x, y in loader:
        x = x.to(device)
        y = y.to(device)
        optimizer.zero_grad()
        outputs = model(x)
        loss = criterion(outputs, y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
        n_batches += 1
    return total_loss / n_batches
```

```python
def evaluate(model, loader, criterion):
    model.eval()
    total_loss = 0.0
    n_batches = 0
    with torch.no_grad():
        for x, y in loader:
            x = x.to(device)
            y = y.to(device)
            outputs = model(x)
            loss = criterion(outputs, y)
            total_loss += loss.item()
            n_batches += 1

    mean_mse = total_loss / n_batches
    rmse = torch.sqrt(torch.tensor(mean_mse)).item()
    return mean_mse, rmse
```

```python

class Dataset_single(torch.utils.data.Dataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __len__(self):
        return len(self.x)

    def __getitem__(self, idx):
        # Convert to numpy first to avoid warnings about dtype/shape
        x_item = torch.tensor(self.x.iloc[idx].to_numpy(), dtype=torch.float32)
        y_item = torch.tensor(self.y.iloc[idx], dtype=torch.float32)

        return x_item, y_item
```

```python
train_dataset = Dataset_single(x_train, y_train)
val_dataset = Dataset_single(X_val, y_val)
test_dataset = Dataset_single(X_test, y_test)

train_dataset[1]
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
input_features = x_train.shape[1]
model = MLP(input_features).to(device)

criterion = nn.MSELoss() # Changed to MSELoss for regression
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# Use the created TensorDatasets with DataLoader
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True) # Reduced batch size for smaller datasets
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)



n_epochs = 10


train_losses = []
val_losses = []
val_rmses = []

print("\n--- Model Training ---")
for epoch in range(n_epochs):
    train_loss = train_one_epoch(model, train_loader, optimizer, criterion)
    val_loss, val_rmse = evaluate(model, val_loader, criterion)
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    val_rmses.append(val_rmse)

    print(f"Epoch {epoch+1}/{n_epochs}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, Val RMSE: {val_rmse:.4f}")

print("\n--- Model Evaluation on Test Set ---")
test_loss, test_rmse = evaluate(model, test_loader, criterion)
print(f"Test Loss: {test_loss:.4f}, Test RMSE: {test_rmse:.4f}")

print(model)

```
